# Schema `vis`
Schema `pipeline.vis` is upstream from `preprocess` and it contains information about the visual stimulus entered by the stimulus software.

```python
%pylab inline
import datajoint as dj
from pipeline import vis
matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)
```

The schema `pipeline.vis` contains information entered by visual stimulus software.  

The main three tables are `vis.Session`, `vis.Condition`, and `vis.Trial`

Each `vis.Session` comprises multiple trials that each has only one condition.  The trial has timing information and refers to the general stimulus condition.

The other manual tables under `vis.Condition` (e.g. `vis.Monet`, `vis.Trippy`, `vis.MovieClipCond`) describe details that are specific to each type of visual stimulus.

The schema also contains some lookup table that cache the more data-intensive stimuli such as noise movies.

```python
dj.ERD(vis).draw()
```

The vis schema refers to the animal information in `pipeline.common` and is then synchronized to the two-photon data by `preprocess.Sync`:

```python
from pipeline import preprocess
(dj.ERD(vis.Session(), locals()) + 1 - 1).draw()
```

## How synchronization works
The stimulus computer and acquisition computers run on different clocks.  The class `preprocess.Sync` matches simultaneously recorded signals in both systems to synchronize the two recordings.

The field `frame_times` in `preprocess.Sync` contains the times of the two-photon frames on the same clock as the `flip_times` in `vis.Trial`.  Furthermore, it has the reference to the `vis.Session`'s `psy.id` and the trial ids of the first and last trial that were played during the given scan (`preprocess.Sync` has 1-to-1 correspondence scans).

```python
preprocess.Sync().show_definition();
```

```python
vis.Trial().show_definition();
```

This way, joining `preprocess.Sync*vis.Trial` provides all the trials that were played for the given scan: 

```python
keys = list(preprocess.Sync().fetch.keys())   # get keys for each trial
matched_trials = preprocess.Sync()*vis.Trial() & 'trial_idx between first_trial and last_trial'
matched_trials & keys[10]    # get all trials for 11th scan in the list
```

But how do you know which what kind of stimulus was displayed?

Each trial has only one stimulus condition.  That stimulus condition is only found in one of the condition tables.  So joining with the wrong kind of condition will yield an empty result and joining with the right table will return exactly one condition for each trial:

```python
matched_trials * vis.Trippy() & keys[10]   # wrong stimulus type
```

```python
matched_trials * vis.Monet() & keys[10]   # right stimulus type
```

In practice, such searching is rarely necessary.  Typically, you first decide what stimulus you want to work with and select the scans that have that stimulus.  This is the principle behind automated population of computed tables.

```python
monet_scans = preprocess.Sync() & matched_trials*vis.Monet() 
monet_scans.proj()   # display
```

```python
monet_scan_keys = list(monet_scans.fetch.keys())  # the keys of all scans that had the monet stimulus
```

These keys can now be used to index scan information and the corresponding trials:

```python
# monet trials with all relevant information for one scan.  Includes stimulus movie.
query = matched_trials*vis.Monet()*vis.MonetLookup() & monet_scan_keys[10]   
```
